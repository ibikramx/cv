```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
 
img = cv2.imread('ab.jpg')
 
alpha = 2
beta = 56
 
result = cv2.addWeighted(img, alpha, np.zeros(img.shape, img.dtype), 0, beta)
 
cv2_imshow(result)
```
