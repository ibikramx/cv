###import necessary library

```
import numpy as np
import matplotlib.pyplot as plt
from skimage import io
```

###load image using skimage

```
picture = io.imread('aa.jpg')
```

###doing operation on image

```
print(picture.shape)
plt.imshow(picture)
```

```
#reverse the image
plt.imshow(picture[::-1])
```

```
#crop the image
crop = picture[200:1100, 300:800]
plt.imshow(crop)
crop.shape
```

```
#add filter to the image
filter = np.where(picture>150, 250, 0)
plt.imshow(filter)
```

```
#landscpe or portrait of image
land = picture[:,:,1].T
plt.imshow(land)
```
